```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("APL_Logistics_preprocessed.csv")
print("Data loaded successfully")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
# Check required columns exist
if 'Order Item Quantity' in df.columns and 'Days for shipment (scheduled)' in df.columns:
    df['Shipping_Pressure_Index'] = df['Order Item Quantity'] / (df['Days for shipment (scheduled)'] + 1)
    print("Shipping_Pressure_Index created")
else:
    print("Columns missing, skipping Shipping_Pressure_Index")

# Inspect
print(df[['Order Item Quantity','Days for shipment (scheduled)','Shipping_Pressure_Index']].head())
```

```python
if 'Shipping Mode' in df.columns:
    df['Shipping_Mode_Express'] = np.where(df['Shipping Mode'] == 'Express', 1, 0)
    print("Shipping_Mode_Express created")
else:
    print("Shipping Mode column missing, skipping Express flag")

# Inspect
print(df[['Shipping Mode','Shipping_Mode_Express']].head())
```

```python
if 'Order Region' in df.columns and 'Order Item Quantity' in df.columns:
    df['Regional_Congestion_Index'] = df.groupby('Order Region')['Order Item Quantity'].transform('count')
    print("Regional_Congestion_Index created")
else:
    print("Columns missing, skipping Regional_Congestion_Index")

# Inspect
print(df[['Order Region','Regional_Congestion_Index']].drop_duplicates().head())
```

```python
if 'Order Item Quantity' in df.columns and 'Order Item Product Price' in df.columns:
    df['Order_Complexity_Score'] = df['Order Item Quantity'] * df['Order Item Product Price']
    print("Order_Complexity_Score created")
else:
    print("Columns missing, skipping Order_Complexity_Score")

# Inspect
print(df[['Order Item Quantity','Order Item Product Price','Order_Complexity_Score']].head())
```

```python
df.to_csv("APL_Logistics_features.csv", index=False)
print("Feature engineering completed. Saved as APL_Logistics_features.csv")
```
