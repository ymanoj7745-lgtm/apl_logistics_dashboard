```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("APL_Logistics.csv", encoding='latin1')  # avoids UnicodeDecodeError
print("Initial shape:", df.shape)
```

```python
# Drop columns not useful for modeling
drop_cols = [
    'Customer Fname', 'Customer Lname', 'Customer Street', 
    'Customer Zipcode', 'Product Name', 'Order Customer Id', 'Customer Id', 
    'Latitude', 'Longitude'
]
df = df.drop(columns=[c for c in drop_cols if c in df.columns])

# Fill missing values
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].fillna('Unknown')

for col in df.select_dtypes(include=np.number).columns:
    df[col] = df[col].fillna(df[col].median())
```

```python
cat_cols = df.select_dtypes(include='object').columns
le_dict = {}
for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    le_dict[col] = le
```

```python
df.to_csv("APL_Logistics_preprocessed.csv", index=False)
print("Preprocessing done, saved as APL_Logistics_preprocessed.csv")
```
